```python
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler,MinMaxScaler,LabelEncoder,OneHotEncoder
```

```python
crypto=pd.read_csv('https://raw.githubusercontent.com/simran200719/crypto/refs/heads/main/bitcoin-sv.csv')
crypto
```

```python
crypto.drop(['Date','Market Cap'],inplace=True,axis=1)
crypto
```

```python
crypto.describe()
```

```python
crypto.isna().sum()
```

```python
crypto['target']=crypto['Close'].shift(-1)
crypto
```

```python
crypto.dropna(inplace=True)
crypto
```

```python
#splitting features and target
x=crypto.iloc[:,:-1]
y=crypto.iloc[:,-1]
print(x)
print(y)
```

```python
train=crypto.iloc[:-157]
test=crypto.iloc[-157:]
print(train)
print(test)
```

```python
scaler_x=StandardScaler()


x_train=train.drop(columns=['target'])
x_test=test.drop(columns=['target'])
x_train_scaled=scaler_x.fit_transform(x_train)
x_test_scaled=scaler_x.transform(x_test)

y_train=train['target']
y_test=test['target']


print(x_train_scaled)
print(x_test_scaled)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train_scaled,y_train)
y_pred=model.predict(x_test_scaled)

print(y_pred)

print(y_test)
```

```python
residual=y_test-y_pred.reshape(-1)
print(residual)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
per_error=((mean_absolute_error(y_test,y_pred))/crypto['target'].mean())*100
r2=r2_score(y_test,y_pred)
print(per_error)
print(r2)
print(crypto['target'].mean())
```
